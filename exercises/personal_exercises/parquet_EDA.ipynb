## Övningar
- 1) Hur ser fördelningen av event-typer ut?
    - Vad är totalen av alla events och vad är procenten för varje event_type

- 2) Hur mycket av din data är faktiskt null?
    - Räkna ut NaN values

- 3) Hur många unika actors finns det, och hur många är bots?
    - Räkna ut antal unika vs antal bots

```python
import pandas as pd


df = pd.read_parquet("part-20260402_080410_592528.parquet")


print(df.shape)
print(df.dtypes)
df.head()
```

```python
# 1) - Hur ser fördelningen av event-typer ut?
# value_counts() räknar förekomster per unikt värde, 
# normalize=True ger procent istället för antal

print(df["event_type"].value_counts())
print("===========================")
print(df["event_type"].value_counts(normalize=True).round(3))
```

```python
# - 2) Hur mycket av din data är faktiskt null?
# - 2) Räkna ut NaN values
#  isnull().sum() räknar NaN per kolumn

df.isnull().sum()
```

```python
# returnerar antal non unique values. .nunique()
total_actors = df["actor_login"].nunique()

# str.contains() söker efter en delsträng i varje värde
bots = df[df["actor_login"].str.contains(r"\[bot\]", na=False)]
bot_actors = bots["actor_login"].nunique()

print(f"Totalt unika actors: {total_actors}")
print(f"Unika bot-actors: {bot_actors}")
print(f"Andel events från bots: {len(bots)/len(df):.1%}")
```
